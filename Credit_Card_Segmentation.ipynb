According to the above outputs we can see that there are 4 types of behaviours in Customers.

             1.Customers who makes only One-Off Purchases.
             2.Customers who makes only Installments Purchases.
             3.Customers who makes only both type of Purchases..
             4.Customers who makes None of the Purchases.
So defining a categorical variable based on the behaviour :-

```python
def purchase(df):
    if ((df['ONEOFF_PURCHASES']>0) & (df['INSTALLMENTS_PURCHASES']==0)):
        return 'One-Off'
    if ((df['ONEOFF_PURCHASES']==0) & (df['INSTALLMENTS_PURCHASES']>0)):
        return 'Installment'
    if ((df['ONEOFF_PURCHASES']>0) & (df['INSTALLMENTS_PURCHASES']>0)):
        return 'Both_Oneoff_Installment'
    if ((df['ONEOFF_PURCHASES']==0) & (df['INSTALLMENTS_PURCHASES']==0)):
        return 'None'
```

```python
df['purchase_type'] = df.apply(purchase,axis=1)
df['purchase_type'].value_counts()
```

### Correlation analysis :-

- Heat map shows that many features are co-related with each other, So i'm going to use dimensionality reduction technique i.e. Principal Component Analysis (PCA).
- PCA transforms a large set of variables into a smaller one that still contains most of the information in the large set. Reducing the number of variables of data.
- Before applying PCA we will standardize data to avoid effect of scale on our result. Centering and Scaling will make all features with equal weight.

```python
col=['ONEOFF_PURCHASES_FREQUENCY','PURCHASES_INSTALLMENTS_FREQUENCY',
     'CASH_ADVANCE_FREQUENCY','PURCHASES_FREQUENCY','PURCHASES_TRX','CASH_ADVANCE_TRX']
df_merge=df_merge[[x for x in df_merge.columns if x not in col ]]
```

```python
df_merge=pd.DataFrame(df_merge)
df_merge.shape
```

### Scaling Data :-

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
df_scaled = sc.fit_transform(df_merge)
```

### Applying PCA :-

```python
from sklearn.decomposition import PCA
```

```python
# We have now 12 features so our n_component will be 12

pc=PCA(n_components=12)
df_pca=pc.fit(df_scaled)
```

```python
# Lets check varience accuracy. Ideally it should be 1 i.e 100%
sum(df_pca.explained_variance_ratio_)
```

```python
var_ratio={}
for i in range(1,13):
    pc=PCA(n_components=i)
    df_pca=pc.fit(df_scaled)
    var_ratio[i]=sum(df_pca.explained_variance_ratio_)
    
var_ratio
```

```python
pca = PCA().fit(df_scaled)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

##### Following by the chart 5 components are giving above 90% variance so we select 5 components.

```python
pc_final=PCA(n_components=5).fit(df_scaled)

df_reduced=pc_final.fit_transform(df_scaled)
```

### Application of Machine Learning algorithm
#### K-Means Clustering :-
##### Finding out K-value using Pseudo F-value (elbow method) :-

```python
from sklearn.cluster import KMeans

#Estimate optimum number of clusters

cluster_range = range( 1, 13 )
Distortions = []

for num_clusters in cluster_range:
    clusters = KMeans( num_clusters )
    clusters.fit( df_reduced )
    Distortions.append( clusters.inertia_ )
    
#Create dataframe with cluster errors
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "Distortions": Distortions } )
```

```python
#Plot line chart to visualise number of clusters
%matplotlib inline  
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.Distortions, marker = "o" )
```

- The elbow diagram shows that the distortion reduces significantly from 4. So, optimal number of clusters could be 4.

##### Finding out K-value using Silhouette Coefficient (SC) :-

```python
from sklearn import metrics
```

```python
# calculate SC for K=2 to K=13

k_range = range(2, 13)

scores = []
for k in k_range:
    km = KMeans(n_clusters=k, random_state=123)
    km.fit(df_reduced)
    scores.append(metrics.silhouette_score(df_reduced, km.labels_))
```

```python
# plot the results

plt.plot(k_range, scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
```

- Based on SC method result is again coming as 4, So let's take K-value as 4 .
